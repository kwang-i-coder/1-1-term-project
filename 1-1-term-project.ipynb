# 2022년 서울 지하철 2호선 역별 승하차 승객 수

## 1.주제 선정 이유:
서울 수도권의 지하철 교통망에서 승객들의 흐림을 찾아 주요 업무 지역의 출퇴근 승객 흐름을 분석하며 더욱 과학적이고 효율적인 지하철 이용 실태 분석 하며 인원분석을 통하여 주변 상권 입지분석을 통해 더 다양한 경로로 발전이 가능하다는 의문이 들어, 이러한 프로젝트를 진행하게 되었다.

## 2.가설 정의:
서울에서 몇 달 정도 생활하면서 지하철 노선도를 분석해본 결과 어딘가를 놀러가기 위하여 지하철을 사용 하는 분들보다는 직장, 학교, 상가를 기준으로 지하철을 사용하는 사람이 굉장히 많았다. 이러한 경험을 바탕으로 생각해 보았을때 아무래도 회사와 학교 지역들이 인구 유입이 굉장히 많을 것이라고 예상한다. 또한, 특정 지역들인 강남 같은 경우에는 직장인들이 거래처와 만남, 본인 직장이 굉장히 많기 때문에 유입이 많을 것이라고 예상되고, 홍대입구, 신촌, 건대입구 같은 경우에는 굉장히 큰 학교와 유명한 대학가가 위치되어 있기 때문에 대학생 + 놀러 오는 사람이 많기 때문에 유입이 많을 것이라고 예상된다.

## 3.인터넷을 통한 데이터 획득:
